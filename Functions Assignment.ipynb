3. What are the different ways to define and call a function in Python ?
   - Standard Function Definition and Call: The def keyword, followed by the function name and parameters, is used to define a function. When calling it, its name is used, followed by arguments enclosed in parenthesis.

   - Function with Default Arguments: When calling a function, functions can be defined with default values for the arguments that are utilized.

   - Functions of Lambda (Anonymous):
  The lambda keyword is used to define a tiny, anonymous function known as a lambda function. Usually, it is employed for brief operations.

   - Function as a parameter: You can pass a function to another function as a parameter.

   - Employing **kwargs and *args: **kwargs permits the passing of a variable number of keyword arguments, whereas *args permits the passing of a variable number of positional arguments.


```python
#question 3 examples
def greet(name):  # Function
    return f"Hello {name}"

class Person:
    def greet(self):  # Method
        return "Hello!"


def greet(name="Guest"):
    return f"Hello {name}"

print(greet())  # Output: Hello Guest
print(greet("Bob"))  # Output: Hello Bob


# Define and call lambda function
add = lambda x, y: x + y
print(add(2, 3))  # Output: 5


def apply_function(func, x, y):
    return func(x, y)

# Define and pass function
result = apply_function(lambda a, b: a * b, 4, 5)
print(result)  # Output: 20


def greet(*names):
    for name in names:
        print(f"Hello {name}")

greet("Alice", "Bob", "Charlie")  # Output: Hello Alice Hello Bob Hello Charlie
```

6. Explain the concept of generators in Python and how they are defined ?
   - Generators: These functions are more memory-efficient since they use the yield keyword to produce values one at a time.


```python
#question 6 example
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
for number in count_up_to(5):
    print(number)
```

7. What are the advantages of using generators over regular functions?
   - Advantages of Generators: Generators are memory-efficient since they produce items one at a time instead of storing the entire result in memory.

```python
#question 7 example
for num in count_up_to(3):  # Only generates one number at a time
    print(num)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - Reduce(), filter(), and map():

    - map() gives each item in an iterable a function.

    - reduce() cumulatively applies a function to an iterable's components.

    - filter() uses a predicate to filter elements from an iterable.


```python
#question 10 example
from functools import reduce
nums = [1, 2, 3]

print(list(map(lambda x: x**2, nums)))
print(reduce(lambda x, y: x + y, nums))
print(list(filter(lambda x: x % 2 == 0, nums)))
```

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list ?
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)
print(sum_of_evens([1, 2, 3, 4, 5, 6]))  # Output: 12 (2 + 4 + 6)
```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string ?
def reverse_string(s):
    return s[::-1]
print(reverse_string("hello"))  # Output: "olleh"
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number ?
def square_numbers(numbers):
    return [num ** 2 for num in numbers]
print(square_numbers([1, 2, 3, 4]))
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200 ?
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print([n for n in range(1, 201) if is_prime(n)])
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms ?
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.a
        else:
            raise StopIteration
fib = FibonacciIterator(5)
print(list(fib))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent ?
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
print(list(powers_of_2(4)))
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple ?
data = [(1, 'apple'), (3, 'cherry'), (2, 'banana')]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit ?
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

temperatures_celsius = [0, 20, 40, 100]
temperatures_fahrenheit = list(map(celsius_to_fahrenheit, temperatures_celsius))
print(temperatures_fahrenheit)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string ?
def remove_vowels(s):
    return ''.join(filter(lambda x: x not in 'aeiouAEIOU', s))
print(remove_vowels("hello world"))  # Output: "hll wrld"
```
